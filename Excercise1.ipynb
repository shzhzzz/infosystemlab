```python
def cleaning_text(text):
    # delete @
    pattern1 = '@'
    text = re.sub(pattern1, '', text)    
    # delete / 
    pattern2 = '[/]'
    text = re.sub(pattern2, '', text)    
    # delete  (
    pattern4 = '[(]' 
    text = re.sub(pattern4, '', text)
    # delete  )
    pattern5 = '[)]' 
    text = re.sub(pattern5, '', text)
    # delete numbers
    pattern6 = '[0-9]' 
    text = re.sub(pattern6, '', text)
    return text
```

```python
from nltk import word_tokenize
#token_words = word_tokenize(d)
#print(token_words)
```

```python
def tokenize_text(text):
  text = re.sub('[.,]', '', text)
  return text.split()
```

```python
from nltk.corpus import wordnet as wn # import lemmatize

def lowerize_word(word):
    # make words lower  e.g., Python =>python
    word=word.lower()
    return word
```

```python
def lemmatize_word(word):
    # lemmatize  e.g., cooked=>cook
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
      return lemma
```

```python
en_stop = nltk.corpus.stopwords.words('english')
print(en_stop)
```

```python
def remove_stopwords(word, stopwordset):
  if word in stopwordset:
    return None
  else:
    return word
```

```python
def remove_characters(word):
    characters = [',', '.', ':', ';', '?', '(', ')', '[', ']', '&', '!', '*', '@', '#', '$', '%','-','...','^','{','}']
    if word in characters:
        return None
    else:
        return word
```

```python
def preprocessing_text(text):
  text = cleaning_text(text)
  tokens = word_tokenize(text)
  tokens = [lowerize_word(word) for word in tokens]
  tokens = [remove_stopwords(word, en_stop) for word in tokens]
  tokens = [word for word in tokens if word is not None]
  tokens = [lemmatize_word(word) for word in tokens]
  tokens = [remove_stopwords(word, en_stop) for word in tokens]
  tokens = [remove_characters(word) for word in tokens]
  tokens = [word for word in tokens if word is not None]
  return tokens
```

```python
preprocessed_docs = [preprocessing_text(text) for text in documents]
preprocessed_docs
```

## BOW

```python
def bow_vectorizer(docs):
  word2id = {}
  for doc in docs:
    for w in doc:
      if w not in word2id:
        word2id[w] = len(word2id)
        
  result_list = []
  for doc in docs:
    doc_vec = [0] * len(word2id)
    for w in doc:
      doc_vec[word2id[w]] += 1
    result_list.append(doc_vec)
  return result_list, word2id
```

```python
bow_vec, word2id = bow_vectorizer(preprocessed_docs)
print(bow_vec)
```

### TF-IDF

```python
def tfidf_vectorizer(docs):
  def tf(word2id, doc):
    term_counts = np.zeros(len(word2id))
    for term in word2id.keys():
      term_counts[word2id[term]] = doc.count(term)
    tf_values = list(map(lambda x: x/sum(term_counts), term_counts))
    return tf_values
  
  def idf(word2id, docs):
    idf = np.zeros(len(word2id))
    for term in word2id.keys():
      idf[word2id[term]] = np.log(len(docs) / sum([bool(term in doc) for doc in docs]))
    return idf
  
  word2id = {}
  for doc in docs:
    for w in doc:
      if w not in word2id:
        word2id[w] = len(word2id)
  
  return [[_tf*_idf for _tf, _idf in zip(tf(word2id, doc), idf(word2id, docs))] for doc in docs], word2id
```

```python
tfidf_vector, word2id = tfidf_vectorizer(preprocessed_docs)
print(tfidf_vector)
print(word2id.items())
```

```python
def simpson_similarity(list_a, list_b):
  num_intersection = len(set.intersection(set(list_a), set(list_b)))
  min_num = min(len(set(list_a)), len(set(list_b)))
  try:
    return num_intersection / min_num
  except ZeroDivisionError:
    if num_intersection == 0:
      return 1.0
    else:
      return 0
```

```python
for i in range(10):
    for j in range(i+1,10):
        print("simpson similarity of document",i,"and",j,"is",simpson_similarity(preprocessed_docs[i],preprocessed_docs[j]))
        
```

```python
def euclidean_distance(list_a, list_b):
  diff_vec = np.array(list_a) - np.array(list_b)
  return np.linalg.norm(diff_vec)
```
